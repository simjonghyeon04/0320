```python

from faker import Faker
import pandas as pd

fake = Faker("ko_KR")

rows = []
for _ in range(1000):
    rows.append({
        "name": fake.name(),
        "address": fake.address(),
        "phone": fake.phone_number(),
        "email": fake.email()

    })

df = pd.DataFrame(rows)
print(df)
df.to_csv("people_1000.csv", index=False, encoding="utf-8")
```

```python
class User:
    def __init__(self, name, location="집"):
        self.name = name
        self.location = location
        self.following = None
        self.follower = None
        self.chase_count = 0

    def following_mode(self, target):
        """Tom이 Jerry를 쫓는 모드"""
        self.following = target.name
        self.follower = self.name
        self.chase_count += 1
        print(f"[{self.chase_count}회차] {self.name}이(가) {target.name}을(를) 쫓습니다!")
        print(f"위치: {self.location} → {target.location}")
        print(f"추격자: {self.follower}, 피추격자: {self.following}")
        print("-" * 30)
tom = User("Tom")
jerry = User("Jerry", "구멍 속")



for i in range(10):
    tom.following_mode(jerry)
```
